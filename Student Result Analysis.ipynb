```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe()   # describes the data as count, mean, minimum, maximum, 25%, 50%, 75% 
```

```python
df.info()   # tells about the data type and non null values in each column
```

```python
df.isnull().sum()   # gives the null value for each column
```

# Drop the unnamed column

```python
df =  df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

## Change weekly study hours cloumn

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10","10")   # changing the weekly hour by 10 from 5-10 
df.head()
```

## Gender Distribution

```python
plt.figure(figsize = (5,5))  # size for the given graph
ax = sns.countplot(data = df, x = "Gender")   # to show the data in bar chart
ax.bar_label(ax.containers[0])   # to give the exact count
plt.title("Gender Distribution")
plt.show
```

```python
# from the above chart we have analysed that the number of females in the data is more than the males.
```

### Relationship between Parent education and student education

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})  
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
# from the above chart we have concluded that the education of the parents have a good impact on their scores
```

### Relationship between Parents marital status over student education

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})  
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

* from the above we have analyzed that the marital status of the parents have no impact or negligible impact on the student score

### Distribution of Ethinc Groups

```python
groupA = df.loc[(df['EthnicGroup']== "group A")].count()
groupB = df.loc[(df['EthnicGroup']== "group B")].count()
groupC = df.loc[(df['EthnicGroup']== "group C")].count()
groupD = df.loc[(df['EthnicGroup']== "group D")].count()
groupE = df.loc[(df['EthnicGroup']== "group E")].count()

l = ["groupA", "groupB","groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
